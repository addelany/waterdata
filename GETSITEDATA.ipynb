```python
import pandas as pd
```

### Variables for each state
Please enter the following variables before running this on any one state. The *mapping set* is the folder in LakeToSiteMapping directory for the particular state. The *statecode* can be obtained from WQP by entering a particular state and checking the state code from the web service calls.

```python
mappingset = "SET5"
state = "MICHIGAN"
countrycode = 'US'
statecode = 'US:26'
```

### Obtai Lake to Site Mappings
Please make sure the path to LakeToSiteMappings is correct. Then read the file to get lake ids and site ids and their mappings

```python
lake_to_site_data = pd.read_csv("sites_extractor_from_water-body-master/notebooks/LakeToSiteMappings/"+ mappingset + "/" + state + "/W2S.csv")
```

```python
lake_to_site = lake_to_site_data[['NHD_LAKE_ID', 'SITE_ID']].copy()
```

```python
lake_to_site
```

```python
site_ids = lake_to_site_data[['SITE_ID']].copy()
```

```python
site_ids
```

```python
from pywqp import pywqp_client
client_instance = pywqp_client.RESTClient()
```

### Obtain raw site data from WQP
The [pywqp REST client](https://github.com/NWQMC/pywqp) is used to get raw site data from WQP. The *params* contain the parameters used to obtain this data.

```python
verb = 'get'
host_url = 'http://waterqualitydata.us'
resource_label = 'result'
params = {'countrycode': countrycode, 'statecode': statecode, 'siteType':'Lake, Reservoir, Impoundment'}
#params = {'countrycode': '', 'statecode': '', 'countycode': '','siteid':site_ids}
equivalent_url = client_instance.create_rest_url(host_url, resource_label, params, mime_type='text/csv')
print equivalent_url
# Make the web-service call and get the response into a DataFrame
result = client_instance.request_wqp_data(verb, host_url, resource_label, params, mime_type='text/csv')
raw_data = client_instance.response_as_pandas_dataframe(result)
```

```python
raw_data
```

### Filter raw site data by excluding sites
Filter the raw site data by excluding the sites not present in the lake to site mappings.

```python
def csv_to_list(df):
    """
        Writes content of csv into list
        Args:
            file_path (str): String
        Returns:
            contents_list (list)
    """
    #df = pd.read_csv(file_path,header=None)
    df.columns = range(df.shape[1])
    contents_list = df[0].tolist()
    return contents_list
```

```python
site_id_list = csv_to_list(site_ids)
#df_filtered = df[df['CharacteristicName'].isin(characteristics)]
```

```python
site_id_list
```

```python
df_filtered = raw_data[raw_data['MonitoringLocationIdentifier'].isin(site_id_list)]
```

```python
df_filtered
```

### Add lake IDs to sites data
The lake ID is included in the filtered site data by merging raw site data with the lake to site data.

```python
final_df = pd.merge(df_filtered, lake_to_site, left_on='MonitoringLocationIdentifier', right_on='SITE_ID', how='left').drop('MonitoringLocationIdentifier', axis=1)
```

```python
final_df
```

### Write site data to file
Site data for each state is written to a separate csv file. Make sure to specify the path of the output files correctly.

```python
final_df.to_csv('data/' + state + '.csv')
```
