### Partition waterbodies into chunks
There are roughly about 100,000 water bodies that belong to each state. Obtaining the NHD-WQP Dataset for such a large dataset is a costly operation. Partitioning this large data into chunks eases the processing of it in later steps. For this, run the methods from partitioning_wb_chunks_size.py . **partitioning_wb_chunks_size** partitions the WB data stored in the previous step into chunks of specified size and stores these data chunks locally in the path *partitionsfolder*. For each chunk, it creates two files - *lakes.csv* which contains data obtained from the NHD shapefile  for lakes in that chunk, and *info.csv* which contains metadata for each chunk.

**write_chunks_path** writes into one csv file the path to all the chunk folders for that state. The last argument to this method needs to be specified as the total number of partitions generated by the **partition_df** method.

```python
partitioning_wb_chunks_size.partition_df(partitionsfolder, state)
```

```python
partitioning_wb_chunks_size.write_chunks_path(partitionsfolder, chunkspathfile, state, 387)
```
